 CORRELATIONS ANALYSIS:

```python
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(16, 6))
mask = np.triu(np.ones_like(corr, dtype=bool))
heat_map = sns.heatmap(corr,
           annot=True,
           mask=mask,
           vmin=-1,
           vmax=1,
           cmap='BrBG')
heat_map.set_title('Triangle Correlations Heatmap', fontdict={'fontsize':20}, pad=20)
plt.show()
```

```python
corr
```

viewing the above anaylsis we come to know that: One of the strong relationships we found is video views and subscribers with a 0.85. Furthermore, we can see a relationship of 0.64 between video views and monthly and annual income. On the other hand, we find an inverse relationship between subscribers and ranking, which is explained by the fact that when a channel has a greater number of subscribers, the ranking decreases.

to get a more deeper idea for the same We will create two new variables that will be the average of the Annual and Monthly income.

```python
df['average_yearly_earnings'] = (df['lowest_yearly_earnings'] + df['highest_yearly_earnings'])/ 2
df['average_monthly_earnings'] = (df['lowest_monthly_earnings'] + df['highest_monthly_earnings'])/ 2
```

finding the Correlation between annual income and video views, uploads, subscribers

```python
ig, axs = plt.subplots(1, 3, figsize=(20,5))

sns.regplot(ax=axs[0], data=df, x='video views', y='average_yearly_earnings', line_kws={'color': 'green'})
sns.regplot(ax=axs[1], data=df, x='uploads', y='average_yearly_earnings', line_kws={'color': 'green'})
sns.regplot(ax=axs[2], data=df, x='subscribers', y='average_yearly_earnings', line_kws={'color': 'green'})

axs[0].grid(True)
axs[0].set_title('Video Views vs Yearly Earnings')
axs[1].grid(True)
axs[1].set_title('Uploads vs Yearly Earnings')
axs[2].grid(True)
axs[2].set_title('Subscribers vs Yearly Earnings')
plt.show()
```

```python
df_corr = df[['average_yearly_earnings', 'video views', 'uploads', 'subscribers']]
corr = df_corr.corr()
sns.heatmap(corr, annot=True, linewidth=0.5, cmap='Reds')
plt.show()
```

from the above graph we can see that It can be seen that annual income has a greater correlation with video view (0.64) than with uploads(0.16) and subscribers(0.53).

```python
country_counts_asc = df['Country'].value_counts(ascending = True)
top_10_countries = country_counts_asc.tail(10)
data = top_10_countries.tolist()
data += [country_counts_asc[:-10].sum()]
labels = top_10_countries.index.tolist() + ['Others']
df_country = pd.DataFrame({'country': labels,
                   'channel': data})
fig_pie = px.pie(df_country, values='channel', names='country')
fig_pie.show()
```

MAPS ANALYSIS:

```python
df_country_map = df['Country'].value_counts(dropna=True)
```

```python
# Converting Series into DataFrames
df_sub_country= pd.DataFrame(df_country_map.index)
df_sub_country = df_sub_country.rename(columns = {0:'Country'})

df_sub_channel = pd.DataFrame(df_country_map.values)
df_sub_channel = df_sub_channel.rename(columns = {0:'Channel'})

df_total_country = pd.concat([df_sub_country, df_sub_channel], axis=1)
print(df_total_country)
```

```python
fig = go.Figure(go.Choropleth(
    locations = df_total_country['Country'],
    locationmode = "country names",
    z = df_total_country['Channel'],
    text = df_total_country['Country'],
    colorscale = 'viridis',
    autocolorscale=False,
    reversescale=True,
    marker_line_color='#efefef',
    marker_line_width=0.5,
    colorbar_ticksuffix = '%',
    colorbar_title = 'Poblacin urbana',
    )
)
fig.show()
```
